# 2-D Line plot

*   Bivariante Analysis
*   categorical -> numerical and numerical -> numerical
*   Use case - Time series data





````python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
````

````python
plt.plot(year,price)
````

````python
# import data from dataset
batsman=pd.read_csv('/sharma-kohli.csv')
batsman
````

````python
plt.plot(batsman['index'],batsman['V Kohli'])
````

````python
 # ploting multiple plot
 plt.plot(batsman['index'],batsman['V Kohli'])
 plt.plot(batsman['index'],batsman['RG Sharma'])
````

````python
# use Lavel & Title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit sharma vs Virat kohli carriar analysis')

plt.xlabel('Seasion')
plt.ylabel('Run scored')
````

````python
# colors(hex) and line(width and style) and marker(size)

plt.plot(batsman['index'],batsman['V Kohli'],color='#F6FF33')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF3377')

plt.title('Rohit sharma vs Virat kohli carriar analysis')

plt.xlabel('Seasion')
plt.ylabel('Run scored')
````

````python
plt.plot(batsman['index'],batsman['V Kohli'],color='#F6FF33',linestyle='solid',linewidth='3',marker='*',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF3377',linestyle='dotted',linewidth=4,marker='o')

plt.title('Rohit sharma vs Virat kohli carriar analysis')

plt.xlabel('Seasion')
plt.ylabel('Run scored')
````

````python
# use legend & location

plt.plot(batsman['index'],batsman['V Kohli'],color='#F6FF33',linestyle='solid',linewidth='3',marker='*',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF3377',linestyle='dotted',linewidth=4,marker='o',label="Rohit")

plt.title('Rohit sharma vs Virat kohli carriar analysis')

plt.xlabel('Seasion')
plt.ylabel('Run scored')

plt.legend(loc='upper left' )
````

````python
#use limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2020)
plt.show()
````

````python
# use grid() function

plt.plot(batsman['index'],batsman['V Kohli'],color='#F6FF33',linestyle='solid',linewidth='3',marker='*',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF3377',linestyle='dotted',linewidth=4,marker='o',label="Rohit")

plt.title('Rohit sharma vs Virat kohli carriar analysis')

plt.xlabel('Seasion')
plt.ylabel('Run scored')

plt.legend(loc='upper left' )

plt.grid()
````

````python
# use show() function

plt.plot(batsman['index'],batsman['V Kohli'],color='#F6FF33',linestyle='solid',linewidth='3',marker='*',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FF3377',linestyle='dotted',linewidth=4,marker='o',label="Rohit")

plt.title('Rohit sharma vs Virat kohli carriar analysis')

plt.xlabel('Seasion')
plt.ylabel('Run scored')

plt.legend(loc='upper left' )

plt.grid()
plt.show()
````

# Scatter plot


*   Bivariate Analysis
*   numerical vs numerical
*   Use case - Finding correlation  



````python
# plt.scatter simple functions

x = np.linspace(-10,10,50)

y=10*x + 3 + np.random.randint(0,300,50)
y
````

````python
plt.scatter(x,y)
````

````python
# plt.scatter on pandas data set
data=pd.read_csv('/batter.csv')
data.head(50)

````

````python
plt.scatter(data['avg'],data['strike_rate'],color='red',marker='*')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
````

````python
# size

tips=sns.load_dataset('tips')
tips
````

````python
# size

plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
````

````python
# scatterplot using plt.plot
# faster
plt.plot(tips['total_bill'],tips['tip'],'o')
````

````python
# plt.plot is faster than plt.scatter
````

# Bar chart

*   Bivariate Analysis
*   Numerical vs Categorical
*   Use case - Aggregate analysis of groups
   



````python
# simple bar chart

children= [10,20,40,10,30]
color = ['red','blue','green','yellow','pink']

plt.bar(color,children,color='yellow')
````

````python
# bar chart using data
````

````python
plt.barh(color,children,color='black')
````

````python
# color and label

df = pd.read_csv('/batsman_season_record.csv')
df
````

````python
plt.bar(df['batsman'],df['2015'])
````

````python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
````

````python
# A problem

children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
````

````python
# Stacked Bar chart

plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
````

# Pie Chart

*  Univariate/Bivariate Analysis
*  Categorical vs numerical
*  Use case - To find contibution on a standard scale



````python
# simple data
data = [23,45,100,20,49]
subjects=['eng','math','hindi','c++','python']

plt.pie(data,labels=subjects)
plt.show()
````

````python
# Pie chart using dataset

df = pd.read_csv('/gayle-175.csv')
df
````

````python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
````

````python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
````

````python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.2,0,0,0,0,0.1],shadow=True)
plt.show()
````
